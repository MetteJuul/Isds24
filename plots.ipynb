```python
import matplotlib.pyplot as plt
import seaborn as sns

# Path to the CSV file
csv_file = './cleaned_with_weight_2.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_file)

# Ensure "FremmÃ¸dte" is numeric
df['FremmÃ¸dte'] = pd.to_numeric(df['FremmÃ¸dte'], errors='coerce')

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))
sns.regplot(y='WMA', x='FremmÃ¸dte', data=df, ax=ax)

# Add labels and title for clarity
ax.set_xlabel('FremmÃ¸dte')
ax.set_ylabel('WMA')
ax.set_title('Trend Line of WMA vs. FremmÃ¸dte')

# Optionally, add gridlines
ax.grid(True)

# Display the plot
plt.show()
```
